```python
#Img encoding code
def encode_img(fname):
    with open("eap.png","rb") as img:
        tmp = base64.b64encode(img.read())
    return tmp 

def decode_new_img(fname):
    tmp = encode_img(fname)
    f1 = open(fname,"wb")
    f1.write(codecs.decode(tmp, "base64"))
    f1.close()
    img = imread(fname)
    return img
```

# Topic Modeling


## Method 1: LDA(Latent Dirichlet Allocation)

## Method 2: NMF(Nonnegative Matrix Factorization)

* Approximation method to get matrix factorization




```python
#Define Helper Ftn to print top words

def print_top_words(model, feature_names, n_top_words):
    for idx, topic in enumerate(model.components_):
        message = "\nTopic #{}:".format(idx)
        message += " ".join([feature_names[i] for i in topic.argsort()][:-n_top_words-1:-1])
        print(message)
        print("=" * 70)


```

우리는 CountVectorizer Class를 Extend.




```python
lemm = WordNetLemmatizer()

class LemmaCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        return lambda doc: (lemm.lemmatize(w) for w in analyzer(doc))       
```

```python
#Storing the entire training text in a list

text = list(train.text.values)

tf_vectorizer = LemmaCountVectorizer(max_df = 0.95,
                                    min_df = 2,
                                    stop_words = "english",
                                    decode_error = "ignore")

tf = tf_vectorizer.fit_transform(text)
```


## 3b. :: LDA(Latent Dirichlet Allocation)


**Unsupervised Learning**

[1] LDA
sklearn's implementation 을 이용할 것.
(gensim의 구현도 있다.)

<br></br>

### <font color = "red"> Important Params </font>

* n_components: # of topics that you specify to the model
* alpha: Dirichlet params that can be linked to the **document topic prior**
* beta: Dirichlet params linked to the **topic word prior**

<br></br>
### Example:
LatentDirichletAllocation 이용.


n_components는 KMeans와 Latent Semantic Analysis Shceme을 이용.






```python
lda = LatentDirichletAllocation(n_components = 11, max_iter = 5,
                                learning_method = "online",
                                learning_offset = 50.,
                                random_state = 0)
```

```python
lda.fit(tf)
```

#### Topics generated by LDA

helper ftn을 써서 각 LDA의 generated topics에 할당된 top 10word 뽑을 것.

```python
n_top_words = 40
print("Topics in LDA model: ")

tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```

```python
first_topic = lda.components_[0]
second_topic = lda.components_[1]
third_topic = lda.components_[2]
fourth_topic = lda.components_[3]
```

```python
first_topic.shape
```

```python

first_topic_words = [tf_feature_names[i] for i in first_topic.argsort()[:-50 -1:-1]]
second_topic_words = [tf_feature_names[i] for i in second_topic.argsort()[:-50 -1:-1]]
third_topic_words = [tf_feature_names[i] for i in third_topic.argsort()[:-50 -1:-1]]
fourth_topic_words = [tf_feature_names[i] for i in fourth_topic.argsort()[:-50 -1:-1]]
```
